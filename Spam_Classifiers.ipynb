# Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

# Loading Dataset

```python
data = pd.read_csv("E:\\Doctorat UQO\\Session Hiver\\Cours Techniques d'Apprentissage\\TP\\Devoir 2\\spambase.data", 
                   names= ['word_freq_make', 'word_freq_address', 'word_freq_all', 'word_freq_3d','word_freq_our',
                           'word_freq_over', 'word_freq_remove','word_freq_internet', 'word_freq_order', 'word_freq_mail',
                           'word_freq_receive', 'word_freq_will', 'word_freq_people','word_freq_report', 'word_freq_addresses',
                           'word_freq_free','word_freq_business', 'word_freq_email', 'word_freq_you','word_freq_credit', 
                           'word_freq_your', 'word_freq_font', 'word_freq_000','word_freq_money', 'word_freq_hp',
                           'word_freq_hpl', 'word_freq_george','word_freq_650', 'word_freq_lab', 'word_freq_labs',
                           'word_freq_telnet','word_freq_857', 'word_freq_data', 'word_freq_415', 'word_freq_85',
                           'word_freq_technology', 'word_freq_1999', 'word_freq_parts','word_freq_pm', 'word_freq_direct',
                           'word_freq_cs', 'word_freq_meeting','word_freq_original', 'word_freq_project', 'word_freq_re',
                           'word_freq_edu', 'word_freq_table', 'word_freq_conference', 'sum_word_freq','char_freq_%3B',
                           'char_freq_%28', 'char_freq_%5B', 'char_freq_%21', 'char_freq_%24', 'char_freq_%23', 'sum_char_freq',
                           'continuous integer', 'Target'])
```

```python
data_x = data.iloc[:,:57]
data_y = data.iloc[:,-1]
```

```python
data_x.shape, data_y.shape
```

```python
data_x = data_x.values
data_y = data_y.values
```

# Splitting the Dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_x, data_y, test_size=0.2, shuffle=True)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

# Standardize features

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

Add some noise for data to train Naive Bayes Classifier

```python
epsi = 0.001

bayes_X_train = X_train + epsi
bayes_X_test = X_test + epsi

bayes_X_train.all() # all() function check if all values are not zero
```

# Create K-Folds for Cross Validation

```python
k__folds = StratifiedKFold(n_splits=5)
```

# KNN Algorithm (k=5)

KNN without Cross Validation

```python
KNN_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski')
KNN_model = KNN_classifier.fit(X_train, Y_train)
```

```python
KNN_pred = KNN_model.predict(X_test)
KNN_acc = accuracy_score(Y_test, KNN_pred)
KNN_error = 1 - KNN_acc
print('KNN Accuracy: ', KNN_acc)
print('KNN Error: ', KNN_error)
```

KNN with Cross Validation

```python
KNN_Cross_Val = cross_val_score(KNN_classifier, X_train, Y_train, cv= k__folds)
KNN_Cross_Val_mean_Error = 1 - (sum(KNN_Cross_Val) / 5)

print('KNN Cross Validation Accuracies: ', KNN_Cross_Val)
print('KNN Cross Validation Mean Error: ', KNN_Cross_Val_mean_Error)
```

# Decision Tree Algorithm

Decision Tree without Cross Validation

```python
DT_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
DT_model = DT_classifier.fit(X_train, Y_train)
```

```python
DT_pred = DT_model.predict(X_test)
DT_acc = accuracy_score(Y_test, DT_pred)
DT_error = 1 - DT_acc
print('Decision Tree Accuracy: ', DT_acc)
print('Decision Tree Error: ', DT_error)
```

Decision Tree with Cross Validation

```python
DT_Cross_Val = cross_val_score(DT_classifier, X_train, Y_train, cv= k__folds)
DT_Cross_Val_mean_Error = 1 - (sum(DT_Cross_Val) / 5)

print('Decision Tree Cross Validation Accuracies: ', DT_Cross_Val)
print('Decision Tree Cross Validation Mean Error: ', DT_Cross_Val_mean_Error)
```

# Naive Bayes Algorithm

Naive Bayes without Cross Validation

```python
NB_classifier = GaussianNB()
NB_model = NB_classifier.fit(bayes_X_train, Y_train)
```

```python
NB_pred = NB_model.predict(bayes_X_test)
NB_acc = accuracy_score(Y_test, NB_pred)
NB_error = 1 - NB_acc
print('Naive Bayes Accuracy: ', NB_acc)
print('Naive Bayes Error: ', NB_error)
```

Naive Bayes with Cross Validation

```python
NB_Cross_Val = cross_val_score(NB_classifier, bayes_X_train, Y_train, cv= k__folds)
NB_Cross_Val_mean_Error = 1 - (sum(NB_Cross_Val) / 5)

print('Naive Bayes Cross Validation Accuracies: ', NB_Cross_Val)
print('Naive Bayes Cross Validation Mean Error: ', NB_Cross_Val_mean_Error)
```

Logistic Regression without Cross Validation

```python
LR_pred = LR_model.predict(X_test)
LR_acc = accuracy_score(Y_test, LR_pred)
LR_error = 1 - LR_acc
print('Logistic Regression Accuracy: ', LR_acc)
print('Logistic Regression Error: ', LR_error)
```

Logistic Regression with Cross Validation

# Support Vector Machine Algorithm

SVM without Cross Validation

```python
SVM_pred = SVM_model.predict(X_test)
SVM_acc = accuracy_score(Y_test, SVM_pred)
SVM_error = 1 - SVM_acc
print('SVM Accuracy: ', SVM_acc)
print('SVM Error: ', SVM_error)
```
